```python
import pandas as pd
pd.set_option('display.max_rows', None) # 展示所有行
pd.set_option('display.max_columns', None) # 展示所有列
import numpy as np
np.set_printoptions(suppress=True)
```

```python
df_sale = pd.read_csv('F_SALES_UN.csv',encoding='utf-16LE')
```

```python
df_sale = df_sale[df_sale['VIP_ID']!=0]
df_sale.head()
```

```python
year_fliter1= [202001]
df_sale1 = df_sale[df_sale['YEAR_MONTH'].isin(year_fliter1)]
df_sale1 = df_sale1.reset_index(drop = True)
df_sale1['VIP_ID']=df_sale1['VIP_ID'].astype(str)
aff= df_sale1.groupby('SKU_CODE')['VIP_ID'].agg(','.join)
aff= aff.apply(pd.unique)
aff= aff.apply(pd.Series)
aff = aff.reset_index()
aff.rename(columns = {0:'vip_id'},inplace = True)
aff.head()
```

```python
def jaccard(list1, list2):
    list1 = list(list1.split(','))
    list2 = list(list2.split(','))
    count = 0
    for i in list1:
        if i in list2:
            count = count + 1
    s1 = count/len(list1)
    s2 = count/len(list2)
    return s1*s2
```

```python
aff_test = aff.head(100)
```

```python
def count_cov(aff):
    df_sim = pd.DataFrame()
    for i in range(len(aff)):
        vip1 = aff.iloc[i,0]
        sku1 = aff.iloc[i,1]
        for j in range(len(aff)):
            vip2 = aff.iloc[j,0]
            sku2 = aff.iloc[j,1]
            if vip1 == vip2:
                sim = 1
            else:
                sim = jaccard(sku1,sku2)
            df_sim.loc[vip1,vip2] = sim
    return df_sim
        
```

```python
test = count_cov(aff_test)
```
